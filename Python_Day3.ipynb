## Retail Inventory Manager


```python
product_categories=('Electronics','Stationery','Clothing')
print(" Categories available =",(product_categories))
```

```python
products_list=['phone','laptop','book','pencil','pants','shirt']
print("Products available =",(products_list))
```

```python
new_product = input("Enter a new product :")
products_list.append(new_product)
print("new product list :",products_list)
```

```python
list(enumerate(products_list))
```

```python
print("Product list with index no :" '\n' ,list(enumerate(products_list)))
```

```python
remove_index = input("Enter the index of the product to remove :")
removed_product = products_list.pop(int(remove_index))
print("Removed product :",removed_product)
print("Final product list :",products_list)
print("Final Categories available =",product_categories)
```
