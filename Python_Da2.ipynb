## Feedback Analyzer

```python
Name = input("Enter your name :")
Feedback = input("Enter your feedback : ")

word_count = len(Feedback.split())
print("Count of words in the sentence :", word_count )

data_count = Feedback.lower().count("data")
print("Count of 'data' word are in the sentence :" ,data_count )

first_5 = Feedback[:5]
last_5 = Feedback[-5:]
print("First 5 characters:", first_5)
print("Last 5 characters:", last_5)

reversed_feedback = Feedback[::-1]
print("Reversed feedback:", reversed_feedback)
```
