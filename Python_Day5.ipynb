## Dictionary, Set, Frozenset

```python
student={"Anu":80,
         "Bala":90,
         "Diana":92,
         "Kanu":85,
         "Janu":89
         }
print(student)
```

```python
student["Dorthy"]=80
student.update({"Sudha":86})
student["Sree"]=80
print(student)
```

```python
hobbies = { "drawing", "dancing", "singing" }
print (hobbies)
```

```python
new_hobbie1 = input("Enter new hobbies: ")
new_hobbie2 = input("Enter new hobbies: ")
new_hobbie3 = input("Enter new hobbies: ")
new_hobbie = {new_hobbie1 ,new_hobbie2, new_hobbie3}
print(hobbies.union(new_hobbie))
```

```python
hobbies_frozen=frozenset(hobbies)
print(hobbies_frozen)
```
