## Contact Directory ,List Methods and Slicing

```python
contacts = [
    ["Bob", "9876543210", "bob@xxx.com"] ,
    [ "Elsa", "9123456780", "elsa@yyy.com"],
    ["Anna", "9988776655", "anna@zzz.com"],
    ["Jeff", "9876543211", "jeff@xxx.com"],
    ["John", "9876543212", "john@xxx.com"]
]
for contact in contacts:
    print(contact)
```

```python
print("Add a new contact")
new_contact = []
new_contact.append(input("Name: "))
new_contact.append(input("Phone: "))
new_contact.append(input("Email: "))
contacts.append(new_contact)
for contact in contacts:
    print(contact)
```

```python
contacts.sort()
for contact in contacts:
    print(contact)
```

```python
for contact in contacts[:3]:
    print(contact)
```

```python
search_name = input("enter the name to search :")
for contact in contacts:
    if search_name.lower() in contact[0].lower():
        print(contact)
```
